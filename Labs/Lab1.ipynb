## Question 1
Part 1:
Here, we have log of income as our x variable and life expectancy as the y. The colors are filtered by world region, and the bubble size is based on population.

```python
#Part 2
import pandas as pd

file = pd.read_csv("Data/q1data.csv")
file

filenew =  file.dropna()
filenew = filenew[filenew['income'] > 0]
file
```

```python
# Part 3
#Making Graph
import plotnine as pt


breaks_labs= [500,1000,2000,4000,8000,16000,32000,64000]
labels_labs = ['500','1000','2000','4000','8000','16k','32k','64k']


(
pt.ggplot(filenew, pt.aes(x= "income", y = "life_exp", fill = "four_regions", size = "population"))
+pt.geom_point()
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.labs(
    title = 'Life Expectancy and Income based on Region and Population',
    x = "Income",
    y= "Life Expectancy",
    fill = "World Regions",
    size = "Population"
)
+pt.scale_x_continuous(trans = 'log10',
breaks = breaks_labs,
labels = labels_labs,
limits = (500,100000)
)
+pt.scale_y_continuous(breaks = [20,30,40,50,60,70,80,90],
labels = ['20','30','40','50','60','70','80','90'],
limits= (20,90))
+pt.theme_light()
+ pt.annotate("text", x=6000, y=50, label="2010", size=100, alpha = .1)
)
```

```python
#Part 4
breaks_labs= [500,1000,2000,4000,8000,16000,32000,64000]
labels_labs = ['500','1000','2000','4000','8000','16k','32k','64k']


(
pt.ggplot(filenew, pt.aes(x= "income", y = "life_exp", fill = "four_regions"))
+pt.geom_boxplot()
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.labs(
    title = 'Life Expectancy and Income based on Region and Population',
    x = "Income",
    y= "Life Expectancy",
    fill = "World Regions",
    size = "Population"
)
+pt.theme_light()
)
```

Using geom_boxplot is useful for seeing an average of each of the regions and how they compare to each other. On the downside, there is not data for every country, and we can't see a difference in populations.

## Question 2
Part 1:
Here, we have exports as our x variable and income as the y. The colors are filtered by world region, and the bubble size is based on energy.

```python
# Part 2
file2 = pd.read_csv("Data/q2data.csv")
file2

filenew2 =  file2.dropna()
filenew2['energy'] = pd.to_numeric(filenew2['energy'], errors='coerce')
```

```python
# Part 3
(pt.ggplot(filenew2, pt.aes(x="exports",y="imports",fill = "four_regions", size = "energy"))
+pt.geom_point()
+pt.labs(
    title = 'Imports and Exports based on different Regions and Energy Usage',
    x = "Exports",
    y= "Imports",
    fill = "World Regions",
    size = "Energy"
)
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.scale_y_continuous(breaks = [50,100,150,200,250,300,350,400],
labels = ['50','100','150','200','250','300','350','400'],
limits= (0,400))
+pt.scale_x_continuous(
breaks = [20,40,60,80,100,120,140,160,180,200,220],
labels = ['20','40','60','80','100','120','140','160','180','200','220'],
limits = (10,220)
)
+pt.theme_light()
+ pt.annotate("text", x=110, y=200, label="1997", size=100, alpha = .1)
)
```

```python
#Part 4
(pt.ggplot(filenew2, pt.aes(x="exports",y="imports",group = "four_regions", color = "four_regions", size = "energy"))
+pt.geom_line(size = 0.5)
+pt.labs(
    title = 'Imports and Exports based on different Regions and Energy Usage',
    x = "Exports",
    y= "Imports",
    fill = "World Regions",
    size = "Energy"
)
+pt.scale_color_manual(values=["aqua","limegreen","red","yellow"])
 +pt.theme_light()
)
```

This plot does an okay job of showing the general trends that imports increase as exports increase. However, it is far too cluttered to see real trends, there are no individual data points, and we also can't factor in energy size.

## Question 3
Part 1:
Here, we have percentage of people using the internet as our x variable and log of GDP/capita as the y. The colors are filtered by world region, and the bubble size is based on income.

```python
# Part 2
file3 = pd.read_csv("Data/q3data.csv")
file3
filenew3 =  file3.dropna()
```

```python
# Part 3
(pt.ggplot(filenew3, pt.aes(x="internet_users",y="gdp",fill = "four_regions", size = "income"))
+pt.geom_point()
+pt.labs(
    title = 'GDP/Capita based on Number of Internet Users \n filtered by Income and World Region',
    x = "Individuals Using the Internet",
    y= "GDP/Capita",
    fill = "World Regions",
    size = "Income"
)
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.scale_y_continuous(trans = 'log10',
    breaks = [200,500,1000,2000,5000,10000,20000,50000,100000],
labels = ['200','500','1000','2000','5000','10k','20k','50k','100k'],
limits= (200,100000))
+pt.scale_x_continuous(
breaks = [10,20,30,40,50,60,70,80,90],
labels = ['10','20','30','40','50','60','70','80','90'],
limits = (0,100)
)
+pt.theme_light()
+ pt.annotate("text", x=50, y=7500, label="2001", size=100, alpha = .1)
)
```

```python
# Part 4
(pt.ggplot(filenew3, pt.aes(x="internet_users",y="gdp",fill = "four_regions", size = "income"))
+pt.geom_violin()
+pt.labs(
    title = 'GDP/Capita based on Number of Internet Users \n filtered by Income and World Region',
    x = "Individuals Using the Internet",
    y= "GDP/Capita",
    fill = "World Regions",
    size = "Income"
)
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.theme_light()
)
```

This violin chart does a good job at showing averages and frequency of values. However, just like the boxplot, we can't see individual data values, and the income variable doesn't show up in our output.
