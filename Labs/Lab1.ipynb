```python
#Part 1
import pandas as pd

file = pd.read_csv("Data/q1data.csv")
file

filenew =  file.dropna()
filenew = filenew[filenew['income'] > 0]
```

```python
#Making Graph
import plotnine as pt


breaks_labs= [500,1000,2000,4000,8000,16000,32000,64000]
labels_labs = ['500','1000','2000','4000','8000','16k','32k','64k']


(
pt.ggplot(filenew, pt.aes(x= "income", y = "life_exp", fill = "four_regions", size = "population"))
+pt.geom_point()
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.labs(
    title = 'Life Expectancy and Income based on Region and Population',
    x = "Income",
    y= "Life Expectancy",
    fill = "World Regions",
    size = "Population"
)
+pt.scale_x_continuous(trans = 'log10',
breaks = breaks_labs,
labels = labels_labs,
limits = (500,100000)
)
+pt.scale_y_continuous(breaks = [20,30,40,50,60,70,80,90],
labels = ['20','30','40','50','60','70','80','90'],
limits= (20,90))
+pt.theme_light()
+ pt.annotate("text", x=6000, y=50, label="2010", size=100, alpha = .1)
)
```

```python


file2 = pd.read_csv("Data/q2data.csv")
file2

filenew2 =  file2.dropna()
filenew2['energy'] = pd.to_numeric(filenew2['energy'], errors='coerce')



```

```python


(pt.ggplot(filenew2, pt.aes(x="exports",y="imports",fill = "four_regions", size = "energy"))
+pt.geom_point()
+pt.labs(
    title = 'Imports and Exports based on different Regions and Energy Usage',
    x = "Exports",
    y= "Imports",
    fill = "World Regions",
    size = "Energy"
)
+pt.scale_fill_manual(values=["aqua","limegreen","red","yellow"])
+pt.scale_y_continuous(breaks = [50,100,150,200,250,300,350,400],
labels = ['50','100','150','200','250','300','350','400'],
limits= (0,400))
+pt.scale_x_continuous(
breaks = [20,40,60,80,100,120,140,160,180,200,220],
labels = ['20','40','60','80','100','120','140','160','180','200','220'],
limits = (10,220)
)
+pt.theme_light()
+ pt.annotate("text", x=110, y=200, label="1997", size=100, alpha = .1)
)


```
